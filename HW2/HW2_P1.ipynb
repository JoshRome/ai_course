#Homework 2: Problem 1
Imports and declaration of data.

```
from mpl_toolkits import mplot3d
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

f1 = lambda x, y: np.sin((x ** 2 + y ** 2))
f2 = lambda x, y: np.cos((x + y))
f3 = lambda x, y: (-x**3 - y**3)
```

```
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)

X, Y = np.meshgrid(x, y)

Z1 = f1(X, Y)
Z2 = f2(X, Y)
Z3 = f3(X, Y)
```

#Minimum of F1
Displays Minimum of Function 1: F1

```
fig = plt.figure()
ax = Axes3D(fig)
ax.contour3D(X, Y, Z1, 100, cmap='cividis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

p1, _ = ax.get_zlim()
i = np.where(Z1 == p1)[0][0]
j = np.where(Z1[i] == p1)[0][0]

ax.plot([X[i,j]], [Y[i,j]], [Z1[i,j]], 'ro', markersize = 5)
plt.tight_layout()

print("Min at Z of {}".format(p1))
```

```
plt.contour(X,Y,Z1, colors='black');
plt.tight_layout()
```

```
plt.contourf(X, Y, Z1, 30, cmap='viridis')
plt.colorbar();
plt.tight_layout()
```

#Maximum of F2
Displays Maximum of Function 2: F2

```
fig = plt.figure()
ax = Axes3D(fig)

ax.contour3D(X, Y, Z2, 100, cmap='RdGy')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

_, p2 = ax.get_zlim()

i = np.where(Z2 == p2)[1][0]
j = np.where(Z2[i] == p2)[0][0]

ax.plot([X[i,j]], [Y[i,j]], [Z2[i,j]], 'ro', markersize = 5)
plt.tight_layout()

print("Max at Z =  {}".format(p2))
```

```
plt.contour(X,Y,Z2, colors='black');
plt.tight_layout()
```

```
plt.contourf(X, Y, Z2, 30, cmap='viridis')
plt.colorbar();
plt.tight_layout()
```

#Function 3 where the gradient is zero.

where its gradient is defined as $[-3x^2,-3y^2]$

If we look at our Point 3, we will see
